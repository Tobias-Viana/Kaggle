### Limpeza e Tratamento de Valores Nulos
Identificação de valores nulos e aplicação de estratégias de imputação.

### Valores nulos do treino

```python
display(treino.isnull().sum())
```

### Valores nulos do teste

```python
display(teste.isnull().sum())
```

```python
cols_with_nulls = [
    'age_first_funding_year', 'age_last_funding_year',
    'age_first_milestone_year', 'age_last_milestone_year'
]

for col in cols_with_nulls:
    median_value_treino = treino[col].median()
    treino[col].fillna(median_value_treino, inplace=True)
    
    teste[col].fillna(median_value_treino, inplace=True)
```

### Codificação de Variáveis Categóricas
Transformação da variável `category_code` em formato numérico.

```python
treino = pd.get_dummies(treino, columns=['category_code'], prefix='category')
teste = pd.get_dummies(teste, columns=['category_code'], prefix='category')

# Garantir que ambos os dataframes tenham as mesmas colunas após o one-hot encoding
train_cols = set(treino.columns)
test_cols = set(teste.columns)

missing_in_test = list(train_cols - test_cols)
for c in missing_in_test:
    if c != 'labels':
        teste[c] = 0

missing_in_train = list(test_cols - train_cols)
for c in missing_in_train:
    treino[c] = 0

test_id = teste['id']

# Garantir que 'labels' não esteja no dataframe de teste antes de reordenar
if 'labels' in treino.columns:
    treino_cols_for_test = treino.drop(columns=['labels']).columns
else:
    treino_cols_for_test = treino.columns

teste = teste[treino_cols_for_test]
teste['id'] = test_id 
```

### Formulação de Hipóteses e Seleção de Features
Com base na análise exploratória, formulamos as seguintes hipóteses e selecionamos as features mais relevantes para o modelo.

#### Hipóteses Formuladas
1.  **Hipótese 1:** "Empresas com maior financiamento total (`funding_total_usd`) e mais rodadas de financiamento (`funding_rounds`) têm maior probabilidade de sucesso."
    *Justificativa:* Maior investimento e múltiplas rodadas de financiamento podem indicar validação de mercado e capacidade de escalar.
2.  **Hipótese 2:** "Empresas com mais relacionamentos (`relationships`) e participantes médios (`avg_participants`) em rodadas de financiamento têm maior probabilidade de sucesso."
    *Justificativa:* Uma rede de contatos robusta e um maior número de participantes em rodadas de financiamento podem sinalizar um ecossistema de apoio mais forte e maior interesse de investidores.
3.  **Hipótese 3:** "Empresas que atingem marcos (`milestones`) mais cedo e com maior frequência (`age_first_milestone_year`, `age_last_milestone_year`) têm maior probabilidade de sucesso."
    *Justificativa:* Atingir marcos importantes de forma consistente demonstra progresso e execução eficaz, fatores críticos para o sucesso de uma startup.


#### Seleção de Features
Com base nas hipóteses e na análise de correlação, as seguintes features foram selecionadas para o modelo:
- `funding_total_usd`: Total de financiamento recebido.
- `funding_rounds`: Número de rodadas de financiamento.
- `relationships`: Número de relacionamentos (investidores, conselheiros, etc.).
- `milestones`: Número de marcos alcançados.
- `avg_participants`: Média de participantes por rodada de financiamento.
- `age_first_funding_year`: Idade da empresa no primeiro financiamento.
- `age_last_funding_year`: Idade da empresa no último financiamento.
- `age_first_milestone_year`: Idade da empresa no primeiro marco.
- `age_last_milestone_year`: Idade da empresa no último marco.
- `is_CA`, `is_NY`, `is_MA`, `is_TX`, `is_otherstate`: Variáveis indicadoras de localização.
- `is_software`, `is_web`, `is_mobile`, `is_enterprise`, `is_advertising`, `is_gamesvideo`, `is_ecommerce`, `is_biotech`, `is_consulting`, `is_othercategory`: Variáveis indicadoras de categoria de negócio (resultantes do One-Hot Encoding).
- `has_VC`, `has_angel`, `has_roundA`, `has_roundB`, `has_roundC`, `has_roundD`: Variáveis indicadoras de tipos de financiamento recebidos.

As features `id` serão removidas, pois não contribuem para a previsão. Todas as colunas categóricas geradas pelo One-Hot Encoding também serão incluídas, pois representam informações importantes sobre o setor da empresa.

```python
# Definindo features (X) e variável alvo (y)
X = treino.drop(columns=["id", "labels"])
y = treino["labels"]

# Preparando X_test
test_ids = teste["id"]
X_test = teste.drop(columns=["id"])

# Alinhar as colunas de X e X_test
train_cols_no_id_label = [col for col in treino.columns if col not in ['id', 'labels']]
test_cols_no_id = [col for col in teste.columns if col not in ['id']]
all_unique_cols = sorted(list(set(train_cols_no_id_label) | set(test_cols_no_id)))
for col in all_unique_cols:
    if col not in X.columns:
        X[col] = 0
    if col not in X_test.columns:
        X_test[col] = 0
X = X[all_unique_cols]
X_test = X_test[all_unique_cols]

print("Features selecionadas (X.columns):")
print(X.columns)
print("Variável alvo (y.name):")
print(y.name)

# Salvar os dataframes X, y e X_test para uso posterior
X.to_csv("X_processed.csv", index=False)
y.to_csv("y_processed.csv", index=False)
X_test.to_csv("X_test_processed.csv", index=False)
```

### Construção e Avaliação do Modelo (RandomForest)
Seleção de um modelo de machine learning adequado e avaliação de seu desempenho com métricas como acurácia, precisão, recall e F1-score.

```python
X = pd.read_csv("X_processed.csv")
y = pd.read_csv("y_processed.csv").squeeze()
X_test = pd.read_csv("X_test_processed.csv")

# Carregar test.csv original para pegar os IDs
test_raw = pd.read_csv("test.csv")
test_ids = test_raw["id"]

# Definir modelo
model = RandomForestClassifier(random_state=42)

# Avaliação com cross-validation para várias métricas
scoring = {
    "accuracy": "accuracy",
    "precision": "precision",
    "recall": "recall",
    "f1": "f1"
}

cv_results = cross_validate(model, X, y, cv=5, scoring=scoring)

# Mostrar resultados médios
print("Acurácia média:", cv_results["test_accuracy"].mean())
print("Precisão média:", cv_results["test_precision"].mean())
print("Recall médio:", cv_results["test_recall"].mean())
print("F1 médio:", cv_results["test_f1"].mean())

# Treinar modelo final em todos os dados
final_model = RandomForestClassifier(random_state=42)
final_model.fit(X, y)

# Prever no conjunto de teste final
y_test_pred = final_model.predict(X_test)

# Gerar arquivo de submissão
submission = pd.DataFrame({
    "id": test_ids,
    "labels": y_test_pred
})
submission.to_csv("submission.csv", index=False)

print("Submissão salva como submission.csv")
```

### Otimização de Hiperparâmetros e Validação Final (RandomForest)
Realização de ajuste fino (finetuning) dos hiperparâmetros do modelo RandomForest para otimizar seu desempenho.

```python
# Carregar os dataframes processados
X = pd.read_csv("X_processed.csv")
y = pd.read_csv("y_processed.csv").squeeze()
X_test_final = pd.read_csv("X_test_processed.csv")

# Dividir os dados de treino em treino e validação
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Definir o espaço de busca para os hiperparâmetros
param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [10, 20, 30, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "bootstrap": [True, False]
}

# Inicializar o modelo RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Inicializar GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring="accuracy")

# Executar a busca pelos melhores hiperparâmetros
grid_search.fit(X_train, y_train)

# Melhores hiperparâmetros e melhor score
print(f"Melhores hiperparâmetros: {grid_search.best_params_}")
print(f"Melhor acurácia (treino com CV): {grid_search.best_score_:.4f}")

# Treinar o modelo final com os melhores hiperparâmetros
best_model = grid_search.best_estimator_

# Fazer previsões no conjunto de validação
y_pred_tuned = best_model.predict(X_val)

# Avaliar o modelo otimizado
accuracy_tuned = accuracy_score(y_val, y_pred_tuned)
precision_tuned = precision_score(y_val, y_pred_tuned)
recall_tuned = recall_score(y_val, y_pred_tuned)
f1_tuned = f1_score(y_val, y_pred_tuned)

print(f"\nResultados do Modelo Otimizado no Conjunto de Validação:")
print(f"Acurácia: {accuracy_tuned:.4f}")
print(f"Precisão: {precision_tuned:.4f}")
print(f"Recall: {recall_tuned:.4f}")
print(f"F1-Score: {f1_tuned:.4f}")
```
